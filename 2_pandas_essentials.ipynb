#### Lesson 3: Indexing, Selection, and Slicing

```python
import pandas as pd
file_path = r"data\full-hospital-data-files\patients.csv"
patients_df = pd.read_csv(
    file_path,
    usecols=["SSN", "BIRTHDATE", "DEATHDATE", "FIRST", "GENDER", "HEALTHCARE_EXPENSES", "HEALTHCARE_COVERAGE"]
)
patients_df
```

```python
# iloc, loc, shorthand
# iloc[row_index, col_index]
patients_df.iloc[0, 0]
```

```python
patients_df.iloc[0, 2:5]
```

```python
type(patients_df.iloc[0, 2:5])
```

```python
patients_df.iloc[:5, 2:5]
```

```python
type(patients_df.iloc[:5, 2:5])
```

```python
# loc[row_label, col_label]
patients_df.loc[0, "BIRTHDATE"]
```

```python
patients_df.loc[:5, "SSN":"GENDER"]
```

```python
# Shorthand
patients_df["BIRTHDATE"]
```

```python
patients_df[["BIRTHDATE", "SSN", "GENDER", "FIRST"]]
```

```python
patients_df[:5]
```

```python
mask = patients_df["HEALTHCARE_EXPENSES"] < 5000
patients_df[mask]["SSN"]
```
