### Module 2: Structured Data (Pandas Essentials)
#### Lesson 1: Series vs. DataFrames

```python
import pandas as pd
```

```python
ages = [22, 25, 64, 21, 30]
age_series = pd.Series(ages)
```

```python
age_series
```

```python
age_series[2]
```

```python
patient_names = ["John", "Jane", "Doe", "Smith", "Emily"]
age_series = pd.Series(ages, index=patient_names)
age_series
```

```python
age_series["Doe"]
```

```python
hospital_data = {
    "Age": [22, 25, 64, 21, 30],
    "Blood Pressure": [120, 130, 140, 110, 125],
    "Room Number": [101, 102, 103, 104, 105]
}

hospital_df = pd.DataFrame(hospital_data, index=patient_names)
hospital_df
```

```python
hospital_df = pd.DataFrame(hospital_data)
hospital_df
```

```python
hospital_df["Age"]
```
