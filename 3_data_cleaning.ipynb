### Module 3: Data Cleaning & Preprocessing
#### Lesson 1: Managing Missing Data (The NaN Problem)

```python
import pandas as pd
import numpy as np

data = {
    'Date': pd.to_datetime(['2026-01-01', '2026-01-02', '2026-01-03', '2026-01-04', '2026-01-05']),
    'Patient_ID': [101, 101, np.nan, 101, 101],
    'Heart_Rate': [72, np.nan, np.nan, 85, np.nan],
    'Notes': [np.nan, "Stable", np.nan, "Recovering", np.nan],
    'Empty_Col': [np.nan, np.nan, np.nan, np.nan, np.nan] # All missing
}

df = pd.DataFrame(data)
print("Original Messy Data:")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.dropna(thresh=3)
```

```python
df.dropna(subset=['Patient_ID', 'Heart_Rate'])
```

```python
df.dropna(axis=1)
```

```python
df.dropna(axis=1, how="all")
```

```python
df
```

```python
df.fillna("Unknown")
```

```python
df.fillna({'Patient_ID': 101, 'Notes': "Unknown"})
```

```python
df
```

```python
df.fillna({'Patient_ID': 101, 'Notes': "Unknown"}, inplace=True)
```

```python
df
```

```python
df.ffill()
```

```python
df.bfill()
```

```python
df.ffill(limit=1)
```
